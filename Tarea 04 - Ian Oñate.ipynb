#### 1. Use el m칠todo de bisecci칩n para encontrar soluciones precisas dentro de $10^{-2}$ para $x^{3}-7x^{2}+14x-6=0$ en cada intervalo

```python
def bisection_method(func, a, b, tol=1e-5):
    """
    Algoritmo de bisecci칩n para determinar una ra칤z de una funci칩n dentro de un intervalo [a, b] con una tolerancia especificada.

    Par치metros:
    func (function): La funci칩n para la cual se busca encontrar una ra칤z.
    a (float): L칤mite inferior del intervalo.
    b (float): L칤mite superior del intervalo.
    tol (float): Precisi칩n deseada para la aproximaci칩n de la ra칤z.

    Retorna:
    float: Una aproximaci칩n de la ra칤z encontrada.
    """
    if func(a) * func(b) >= 0:
        print("La funci칩n no presenta un cambio de signo en el intervalo proporcionado.")
        return None

    c = a
    while (b - a) >= tol:
        c = (a + b) / 2
        if func(c) == 0.0:
            break
        if func(c) * func(a) < 0:
            b = c
        else:
            a = c
    return c
```

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6
```

$a. [0, 1]$

```python
a, b = 0, 1
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

$b. [1, 3.2]$

```python
a, b = 1, 3.2
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

$c. [3.2, 4]$

```python
a, b = 3.2, 4
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

#### Suponga que $L = 10 cm, r = 1 cm$ y $V =12.4 cm$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 cm$ 

```python
def V(h, L=10, r=1):
    return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h/r) - h * np.sqrt(r**2 - h**2)) - 12.4

a, b = 0, 2  # Intervalo inicial
root = bisection_method(V, a, b, tol=0.01)
print(f"La profundidad del agua es: {root} cm")
```

#### 2. Un objeto que cae verticalmente a trav칠s del aire est치 sujeto a una resistencia viscosa, as칤 como a la fuerza de gravedad. Suponga que un objeto con masa $m$ cae desde una altura $s_0$ y que la altura del objeto despu칠s de $t$ segundos es $$s(t)=s_0 - \frac{mg}{k}t + \frac{m^2g}{k^2}(1-e^{-kt/m}),$$ donde $g=9.81m/s^2$ y $k$ representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $s_0=300m$, $m=0.25kg$ y $k=0.1Ns/m$. Encuentre dentro de $0.01\space segundos$, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
g = 9.81
m = 0.25
k = 0.1
s0 = 300

def s(t):
    return s0 - (m/k)*g*t + (m/k)**2 * g * (1 - np.exp(-(k/m)*t))

def height(t):
    return s(t)

a, b = 0, 100  # Intervalo inicial grande
root = bisection_method(height, a, b, tol=0.01)
print(f"El tiempo que tarda en golpear el piso es: {root} segundos")
```

### 1. Use el teorema 2.1 para encontrar una cota para el n칰mero de iteraciones necesarias para lograr una aproximaci칩n con precisi칩n de 10^-4 para la soluci칩n de 洧논^3  洧논  1 = 0 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaci칩n para la ra칤z con este grado de precisi칩n. 


```python
import numpy as np

def bisection_method(func, a, b, tol=1e-4):
    """
    M칠todo de bisecci칩n para encontrar una ra칤z de una funci칩n en un intervalo [a, b] con una tolerancia tol.
    """
    if func(a) * func(b) >= 0:
        print("La funci칩n no cambia de signo en el intervalo dado.")
        return None

    c = a
    while (b - a) >= tol:
        c = (a + b) / 2
        if func(c) == 0.0:
            break
        if func(c) * func(a) < 0:
            b = c
        else:
            a = c
    return c

# Funci칩n f(x) = x^3 - x - 1
def f(x):
    return x**3 - x - 1

# Intervalo [1, 2] y tolerancia 10^-4
a, b = 1, 2
tol = 1e-4

# C치lculo del n칰mero de iteraciones necesarias
n = int(np.ceil(np.log2((b - a) / tol)))
print(f"N칰mero de iteraciones necesarias: {n}")

# Aproximaci칩n de la ra칤z
root = bisection_method(f, a, b, tol)
print(f"La ra칤z aproximada es: {root}")
```

### 2. La funci칩n definida por 洧녭(洧논) = sin 洧랢 洧논 tiene ceros en cada entero. Muestre cuando 1 < 洧녩 < 0 y 2 < 洧녪 <3, el m칠todo de bisecci칩n converge a: a. 0, 洧멇롐 洧녩 + 洧녪 < 2 b. 2, 洧멇롐 洧녩 + 洧녪 > 2 c. 1, 洧멇롐 洧녩 + 洧녪 = 2

```python
def g(x):
    return np.sin(np.pi * x)

# Intervalos y tolerancia
intervals = [(-1, 0), (2, 3)]
tol = 1e-4

# Aplicar el m칠todo de bisecci칩n para cada intervalo
for a, b in intervals:
    root = bisection_method(g, a, b, tol)
    print(f"La ra칤z aproximada en el intervalo [{a}, {b}] es: {root}")

# An치lisis de eficiencia
a, b = -1, 0
if a + b < 2:
    print("El rango converge a 0 porque a + b < 2.")
elif a + b > 2:
    print("El rango converge a 2 porque a + b > 2.")
else:
    print("El rango converge a 1 porque a + b = 2.")
```
